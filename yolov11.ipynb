```python
file = r'C:\Users\johna\Desktop\CS3264proj\Coffee_room_01\Coffee_room_01\Videos\video (1).avi'
annotation_file = r'C:\Users\johna\Desktop\CS3264proj\Coffee_room_01\Coffee_room_01\Annotation_files\video (1).txt'
output_frame = r'C:\Users\johna\Desktop\CS3264proj\data\train\images'
output_annotation = r'C:\Users\johna\Desktop\CS3264proj\data\train\labels'
```

```python
#extract from annotations number of frames and frames that are falling
fall = []
total_frames = 0
with open(annotation_file, 'r') as f:
    lines = f.readlines()
    lines[0] = lines[0].strip() 
    fall.append(lines[0])
    lines[1] = lines[1].strip()
    fall.append(lines[1])
    for i in range(2, len(lines)):
        lines[i] = lines[i].strip()
        line = lines[i].split(',')
        print(line)
        x_min = int(line[2])
        x_max = int(line[3])
        y_min = int(line[4])
        y_max = int(line[5])
        #find centre and wifhgt and heihgt YOLO format
        x_centre = abs((x_min + x_max) / 2)
        y_centre = abs((y_min + y_max) / 2)
        width = x_max - x_min
        height = y_max - y_min
    
        with open(output_annotation + '1_coffeerm1.txt', 'a') as f:
            if line[0] == fall[0] or line[0] == fall[1]:
                object_class = 1
                f.write(f"{object_class} {x_centre} {y_centre} {width} {height}\n")
                total_frames += 1   
            else:
                object_class = 0
                f.write(f"{object_class} {x_centre} {y_centre} {width} {height}\n")
                total_frames += 1
            continue


    
```
